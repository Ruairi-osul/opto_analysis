```python
from spiketimes.simulate import imhomogenous_poisson_process
from spiketimes.plots import raster_multiple_neurons, raster_single_neuron
from spiketimes.simulate import shuffled_isi_spiketrains, jitter_spiketrains
import matplotlib.pyplot as plt
```

```python
st = imhomogenous_poisson_process([(3, 5)])
ax = raster_single_neuron(st)
ax.set_title("Original_neuron")
```

```python
st2 = jitter_spiketrains(st, n=10, sampling_interval=0.3)
st3 = shuffled_isi_spiketrains(st, n=10)
```

```python
_, ax = plt.subplots(ncols=2, figsize=(10, 5))

ax[0] = raster_multiple_neurons(st2, ax=ax[0])
ax[0].set_title("Jittered Spiketimes")

ax[1] = raster_multiple_neurons(st3, ax=ax[1])
ax[1].set_title("Shuffled Inter-Spike-Intervals")
```
